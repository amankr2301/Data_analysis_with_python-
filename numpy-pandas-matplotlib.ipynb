##                             Line Plot

```python
x = np.arange(0,10)
y = np.square(x)

plt.plot(x , y)
plt.xlabel("Numbers")
plt.ylabel("squares")
plt.title("Basic Line Plot")
plt.show()
```

```python
## Creating a customized line plot 

plt.plot(x , y , color = "red" , linestyle = "--")
## to create a grid 
plt.grid()
```

```python
## Multiple plots 

x = np.arange(1,6) ; 
y = np.square(x) ; 

plt.figure(figsize = (9,5))

## (2,2,1) means there will be 2 rows 2 columns and 1 is the index of this particular subplot
plt.subplot(2,2,1) 
plt.plot(x,y , color = "green")

plt.subplot(2,2,2)
plt.plot(np.sqrt(x) , y , color = "red")


plt.subplot(2,2,4)
plt.plot(np.square(x) , y , color = "blue")

plt.subplot(2,2,3)
plt.plot(np.exp(x) , y , color = "orange")
```

## BAR PLOT 

```python
categories = ['A' , 'B' , 'C' , 'D' , 'E']
value = [5,11,4,9,3]

plt.bar(categories , value , color = 'purple')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

## Pie plots 

```python
labels = ['A' , 'B' , 'C' , 'D']
sizes = [30 , 20 , 40 , 10]
explode = [0.2 , 0 , 0 , 0]
colors = ['gold' , 'yellowgreen' , 'lightcoral' , 'lightskyblue']

plt.pie(sizes ,explode = explode,labels = labels , colors = colors , shadow = True , autopct = "%1.1f%%")
```

## Seaborn 

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
##tip = sns.load_dataset(df)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
## Creating a scatter plot 

sns.scatterplot(x = 'total_bill' , y = 'tip' , data = tips)
plt.title("Scatter plot for total_bill vs the tip")
plt.show()
```

```python
## Line plot 
sns.lineplot(x = 'size' , y = 'total_bill' , data = tips , marker='x')
plt.title("Line plot of size vs total_bill")
```

```python
sns.scatterplot(x = 'size' , y = 'total_bill' , data = tips)
```

```python
## Categorical plot 
sns.barplot(x = 'day' , y = 'total_bill' , data=tips)
```

```python
## Box plot 
sns.boxplot(x = 'day' , y='total_bill' , data = tips)
```

```python
## Violin plot 
sns.violinplot(x = 'day' , y = 'total_bill' , data = tips)
```

```python
sns.histplot(tips['total_bill'] , bins=10 , kde = True)
```

```python
## if we just want the kde that is the shape from the hist plot 
sns.kdeplot(tips['total_bill'] , fill = True)
```

```python
sns.pairplot(tips)
```

```python
## correlation 
corr =tips[['tip' , 'size' , 'total_bill']].corr()
corr
```

```python
# Heatmap just a way to visualize this correlation 
sns.heatmap((corr) , annot = True , cmap = 'coolwarm')
```
